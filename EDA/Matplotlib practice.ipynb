```python
%matplotlib inline
plt.rcParams['figure.figsize']= 8,4
```

```python
x= np.linspace(0,10,100)
```

```python
fig= plt.figure()
```

```python
plt.plot(x, np.sin(x), '--')

plt.plot(x, np.cos(x), '-')
```

```python
# creating sub plot
```

```python
plt.figure()

plt.subplot(2,1,1)                   #(2rows,1column,1 panel number)
plt.plot(x,np.sin(x))                


plt.subplot(2,1,2)                  #(2rows,1column,2 panel number)
plt.plot(x,np.cos(x))
```

```python
x1= np.linspace(0,2.5,100)

plt.plot(x1, x1, label = 'Linear')
plt.plot(x1, x1**2, label = 'Quadratic')
plt.plot(x1, x1**3, label = 'Cubic')

plt.xlabel('x axis')
plt.ylabel('y axis')

plt.title('Simple plot')

plt.legend()
plt.show()
```

```python
# Object oriented API plot
fig, ax=plt.subplots(2)

ax[0].plot(x, np.sin(x),'r')
ax[1].plot(x, np.cos(x),'g');
```

```python
fig = plt.figure()

x2 = np.linspace(0, 5, 10)
y2 = x2 ** 2

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x2, y2, 'r')

axes.set_xlabel('x2')
axes.set_ylabel('y2')
axes.set_title('title')
```

```python
fig= plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

fig.tight_layout()              # set spacing between the subplot , we can also use pad=5 in the function

ax1.set_title('Plot 1')
ax2.set_title('Plot 2')
ax3.set_title('Plot 3')
ax4.set_title('Plot 4')
```

# scatter plot

```python
x6 = np.linspace(1,10,30)               # scatter plot use plt.plot or ax.scatter 

y6 = np.sin(x6)

plt.plot(x6,y6,'o', color = 'black')
```

# histogram

```python
data= np.random.randn(1000)     # its consider frequency
plt.hist(data)
plt.show()
```

# Bar chart

```python
data1 = [5,15,50,25]

plt.bar(range(len(data1)),data1)

plt.show()
```

# Horizontal bar chart

```python
plt.barh(range(len(data1)), data1)

plt.show()
```

# Error bar chart

```python
x9 = np.arange(0, 4, 0.2)

y9 = np.exp(-x9)  

e1 = 0.1 * np.abs(np.random.randn(len(y9)))

plt.errorbar(x9, y9, yerr = e1, fmt = '.-')

plt.show()
```

# Stacked bar chart

```python
a = [15,20,50,70]           # we give two inputs it consider top and bottom

b= [15,30,45,60]

z= range(4)

plt.bar(z,a, color='r')
plt.bar(z,b, color='g', bottom=a)
```

# Pie chart

```python
plt.figure(figsize=(7,7))

x10 = [30,40,10,20]

plt.pie(x10, labels= ('computer','electronics','civil','mechancial'))

plt.show()
```

# Box plot

```python
data2 = np.random.randn(100)

plt.boxplot(data2)

plt.show()
```

# Area chart

```python
a1 = range(1,6)
b1= [1,3,8,6,5]
plt.fill_between(a1,b1)    # consider (x,y) argument

plt.show()
```

```python
plt.stackplot(a1,b1)                      # use this function also
```
