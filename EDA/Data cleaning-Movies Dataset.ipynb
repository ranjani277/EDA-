```python
df['MOVIES']
```

```python
df['MOVIES'] = df['MOVIES'].str.replace('r\W',' ')
```

```python
df['MOVIES'].head()
```

```python
df['YEAR']
```

```python
df['YEAR'] = df['YEAR'].str.replace(r'[\W_\(\)-]',' ')
```

```python
df['YEAR']
```

```python
df['GENRE'] = df['GENRE'].str.replace(r'[\n ,]',' ')
```

```python
df['GENRE']
```

```python
df['ONE-LINE']
```

```python
df['ONE-LINE'] = df['ONE-LINE'].str.replace(r'[\n]','')
```

```python
df['ONE-LINE']
```

```python
df.rename(columns={'STARS':'CAST&CREW'},inplace = True)
```

```python
df.columns
```

```python
df['CAST&CREW'] = df['CAST&CREW'].str.replace(r'[\n]','')
```

```python
df['CAST&CREW']
```

```python
df.info()
```

```python
df.head()
```

```python
df['VOTES']=df['VOTES'].str.replace(',','')
```

```python
clean_data = df.copy()
```

```python
clean_data.isnull().sum()
```

```python
# Dealing with Missing value
```

```python
clean_data['RATING'] = clean_data['RATING'].fillna(np.mean(pd.to_numeric(clean_data['RATING']))).round(2)
```

```python
clean_data
```

```python
clean_data['VOTES'] = clean_data['VOTES'].astype(float)
```

```python
clean_data['VOTES'] = clean_data['VOTES'].fillna(np.mean(pd.to_numeric(clean_data['VOTES'])))
```

```python
clean_data.head()
```

```python
clean_data['RunTime'] = clean_data['RunTime'].fillna(np.mean(pd.to_numeric(clean_data['RunTime']))).round(2)
```

```python
clean_data.head()
```

```python
clean_data.info()
```

```python
clean_data.isnull().sum()
```

```python
columns = clean_data.columns
for i in columns:
    if  clean_data[i].dtype==('object'):
         clean_data[i]=clean_data[i].astype('category')
    
```

```python
clean_data.info()
```

```python
file_path = r"C:\Users\user\Cleandata.csv"
clean_data.to_csv(file_path,index=False)
```
