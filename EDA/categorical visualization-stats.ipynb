# Categorical data visualization

```python
import pandas as pd
import numpy as np
```

```python
data = {'Categery' : ['Audi','BMW','Mercedes'] ,
        'Frequency': [128,98,113]}
cat = pd.DataFrame(data)
```

```python
cat
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Visualization 
```

```python
vis1 = sns.barplot(cat, x= 'Categery', y= 'Frequency',palette=('green','red','blue'))
for i in vis1.containers:
    vis1.bar_label(i)
```

```python
vis2= cat.groupby(('Categery')).sum().plot(kind='pie',y= 'Frequency', autopct='%1.0f%%')
```

```python
plot = cat.plot.pie(y='Frequency', title="Car shop ", legend=False, autopct='%1.1f%%',shadow=True )          
```

# Another categorical dataset

```python
cat1 = pd.DataFrame({'Categery' : ['New Yord','LA','San Francisco'] ,
        'Frequency': [12374,17456,19965]})
```

```python
cat1
```

```python
vis3= sns.barplot(cat1,x= 'Categery',y='Frequency', palette = ('red','purple','pink'))
for i in vis3.containers:
    vis3.bar_label(i)
```

```python
plot1 = cat1.plot.pie(y='Frequency', title="Ice cream sales", legend=False,autopct='%1.1f%%', shadow=True, )  
```

```python
cat1["cumpercentage"] = cat1["Frequency"].cumsum()/cat1["Frequency"].sum()*100
```

# Cumulative frequency plot(pareto Diagram)

```python
fig,ax = plt.subplots()
ax.bar(cat1.Categery,cat1['Frequency'],color='blue')
ax2 = ax.twinx()
ax2.plot(cat1.Categery,cat1['cumpercentage'],color='C1',marker='D',ms=7)
ax.tick_params(axis="y", colors="C0")
ax2.tick_params(axis="y", colors="C1")
plt.show()
```
